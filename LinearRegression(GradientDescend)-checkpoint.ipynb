```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression

import sympy as sp
from sympy import symbols
from sympy import Derivative
from sympy.plotting import plot
from sympy.plotting import plot3d

from IPython.display import display
```

```python
sins = np.sin(np.arange(0, 10, 0.1))
result = sins + np.random.random(100) + np.arange(100)/20 + 3

RESULT_NAME = 'result'
INDEX_NAME = 'index'

df = pd.DataFrame({RESULT_NAME: result})
df.head(3)
```

```python
df[INDEX_NAME] = df.index.to_numpy() + 1
df[INDEX_NAME] += np.random.normal(0, 4, size=100)#np.vectorize(lambda t: -1 if t==1 else 1)(np.random.randint(1, 3, 100))*4

df[[INDEX_NAME]] = MinMaxScaler().fit_transform(df[[INDEX_NAME]])

df[INDEX_NAME].corr(df[RESULT_NAME])
```

```python
sns.scatterplot(data=df, x=INDEX_NAME, y=RESULT_NAME);
```

```python
lr = LinearRegression()
lr.fit(df[[INDEX_NAME]], df[RESULT_NAME])

sns.scatterplot(data=df, x=np.arange(len(df)), y=lr.predict(df[[INDEX_NAME]]));
sns.scatterplot(data=df, x=np.arange(len(df)), y=RESULT_NAME);
```

```python
x0, x1, x2, x3 = sp.symbols('x0 x1 x2 x3')
```

```python
X_mat = sp.Matrix(df[INDEX_NAME].to_numpy())
X_mat
```

```python
sinXx1_mat = X_mat.applyfunc(lambda x: sp.sin(x1*x))
sinXx1_mat
```

```python
sinXx1_mat
```

```python
sinXx1_mat = sinXx1_mat.col_insert(sinXx1_mat.shape[1], sp.ones(sinXx1_mat.shape[0], 1))
sinXx1_mat = sinXx1_mat.col_insert(sinXx1_mat.shape[0], sp.Matrix(df[INDEX_NAME].to_numpy()))

b_vec = sp.Matrix([[x0, x2, x3]]).T

y_vec = sp.Matrix(df[RESULT_NAME].to_numpy())

display(sinXx1_mat, b_vec, y_vec)
```

```python
Xb_mat = sinXx1_mat@b_vec
Xb_mat
```

```python
y_Xb_matr = (Xb_mat-y_vec)
y_Xb_matr
```

```python
y_Xb_square_mat = y_Xb_matr.T * y_Xb_matr#.applyfunc(lambda x: x**2)
y_Xb_square_mat
```

```python
y_Xb_square_mat.shape[0]
```

```python
mse = ((y_Xb_square_mat)/y_Xb_matr.shape[0])[0]
mse
```

```python
#mse.simplify()
```

```python
mse.subs({'x0':1.5, 'x1':4, 'x2': 3})
```

```python
x0x1x2x3 = sp.Matrix([x0, x1, x2, x3])
x0x1x2x3
```

```python
x0x1x2x3_dif = sp.diff(mse, x0x1x2x3)
x0_dif = x0x1x2x3_dif[0]
x1_dif = x0x1x2x3_dif[1]
x2_dif = x0x1x2x3_dif[2]
x3_dif = x0x1x2x3_dif[3]
```

```python
coefs = {'x0':0, 'x1':10, 'x2':0, 'x3':0}
```

```python
LEARNING_RATE = 0.1

for i in range(100):
    
    coefs['x0'] -= float(x0_dif.subs(coefs)) * LEARNING_RATE   
    coefs['x1'] -= float(x1_dif.subs(coefs)) * LEARNING_RATE   
    coefs['x2'] -= float(x2_dif.subs(coefs)) * LEARNING_RATE
    coefs['x3'] -= float(x3_dif.subs(coefs)) * LEARNING_RATE

    
mse_result = mse.subs(coefs)


result_func = coefs['x0']*sp.sin(coefs['x1']*x0) + coefs['x2'] + coefs['x3']*x0


display(coefs, mse_result, result_func)

plot(result_func, (x0, 0, 1));
```

```python
l = np.arange(0, 1, 0.01)
result_l = []

for i in l:
    result_l.append(result_func.subs({'x0':i}))

    
    
    
sns.scatterplot(x=l, y=result_l);
sns.scatterplot(data=df, x=INDEX_NAME, y=RESULT_NAME);
```
